# Exercise: Decision Trees in sklearn

In this section, you'll use decision trees to fit a given sample dataset.

Before you do that, let's go over the tools required to build this model.

For your decision tree model, you'll be using scikit-learn's <code>Decision Tree Classifier</code> class. This class provides the functions to define and fit the model to your data.

```
>>> from sklearn.tree import DecisionTreeClassifier
>>> model = DecisionTreeClassifier()
>>> model.fit(x_values, y_values)
```

In the example above, the <code>model</code> variable is a decision tree model that has been fitted to the data <code>x_values</code> and <code>y_values</code>. Fitting the model means finding the best tree that fits the training data. Let's make two predictions using the model's <code>predict()</code> function.

```
>>> print(model.predict([ [0.2, 0.8], [0.5, 0.4] ]))
[[ 0., 1.]]
```

The model returned an array of predictions, one prediction for each input array. The first input, <code>[0.2, 0.8]</code>, got a prediction of <code>0.</code>. The second input, <code>[0.5, 0.4]</code>, got a prediction of <code>1.</code>.

## Hyperparameters

When we define the model, we can specify the hyperparameters. In practice, the most common ones are

- <code>max_depth</code>: The maximum number of levels in the tree.
- <code>min_samples_leaf</code>: The minimum number of samples allowed in a leaf.
- <code>min_samples_split</code>: The minimum number of samples required to split an internal node.

For example, here we define a model where the maximum depth of the trees <code>max_depth</code> is 7, and the minimum number of elements in each leaf <code>min_samples_leaf</code> is 10.

```
python >>> model = DecisionTreeClassifier(max_depth = 7, min_samples_leaf = 10)
```
