$$
E(S) = \sum_{i=1}^c -p_i \log_2 p_i
$$

```python
# Import libraries necessary for this quiz
import numpy as np
import pandas as pd
from IPython.display import display # Allows the use of display() for DataFrames

# Load the dataset
# URL for our dataset, ml-bugs.csv
URL = "https://drive.google.com/file/d/1_D8Maq3hFcCDZdO7mVodfGlMInj26n9i/view?usp=sharing"
FILE_PATH = "https://drive.google.com/uc?export=download&id=" + URL.split("/")[-2]

df = pd.read_csv(FILE_PATH)

display(df.head())
```

### Calculate Entropy of the Root Node

```python
# TO DO
df.groupby(["Species"])["Species"].count()
```

```python
# TO DO
prop_lobug = 14 / (14 + 10)
prop_mobug = 10 / (14 + 10)

entropy_root = -prop_lobug*np.log2(prop_lobug) - prop_mobug*np.log2(prop_mobug)
entropy_root
```

### Calculate Entropy of the Child Nodes (Split by Blue Colour)

```python
# TO DO
df.groupby(["Color", "Species"])["Species"].count()
```

```python
# TO DO
entropy_blue = -6/10 * np.log2(6/10) - 4/10 * np.log2(4/10)
entropy_not_blue = -8/14 * np.log2(8/14) - 6/14 * np.log2(6/14)

entropy_color = 10/24 * entropy_blue + 14/24 * entropy_not_blue
entropy_color
```

```python
# Information Gain
entropy_root - entropy_color
```
