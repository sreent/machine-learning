# Exercise-Quiz: Absolute and Square Trick

## Quiz for Absolute Trick

Let's say that we have a line whose equation is **y = -0.6x + 4**. For the point **(x,y) = (-5, 3)**, apply the **absolute trick** to get the new equation for the line, using a learning rate of $alpha=0.1$.

Report your answer in the form **y = w_1x + w_2**, substituting appropriate values for **w_1** and **w_2**.

Note: $y = (w_1+p\alpha)x + (w_2+α)$

```python
The new equation of the line should be y = −0.1x+3.9
```

## Quiz for Square Trick

Let's say that we have a line whose equation is **y = -0.6x + 4**. For the point **(x,y) = (-5, 3)**, apply the **square trick** to get the new equation for the line, using a learning rate of $alpha=0.1$.

Report your answer in the form **y = w_1x + w_2**, substituting appropriate values for **w_1** and **w_2**.

Note: $y = (w_1 + p(q-q^{'})\alpha)x + (w_2+(q-q^{'})α)$

```python
The new equation of the line should be y=−0.4x+3.96
```
